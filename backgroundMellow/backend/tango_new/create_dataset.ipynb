```python
"""
Create train_mydataset.json from DATASET/ directory.
Structure: DATASET/<audio_class>/<file>.wav
Each line: {"dataset": "my_dataset", "location": "<abs_path>", "captions": "<audio_class>"}
"""
import json
from pathlib import Path

# Configuration
DATASET_DIR = Path("../../DATASET")  # relative to notebook location when run from BTP/
OUTPUT_FILE = Path("tango_new/data/train_mydataset.json")
DATASET_NAME = "my_dataset"
AUDIO_EXTENSIONS = {".wav", ".mp3", ".flac", ".ogg"}

def create_train_json(dataset_dir: Path, output_path: Path, dataset_name: str = "my_dataset"):
    dataset_dir = Path(dataset_dir).resolve()
    if not dataset_dir.is_dir():
        raise FileNotFoundError(f"DATASET directory not found: {dataset_dir}")

    entries = []
    for audio_class_dir in sorted(dataset_dir.iterdir()):
        if not audio_class_dir.is_dir():
            continue
        class_name = audio_class_dir.name
        for audio_file in sorted(audio_class_dir.iterdir()):
            if audio_file.suffix.lower() in AUDIO_EXTENSIONS:
                entries.append({
                    "dataset": dataset_name,
                    "location": str(audio_file.resolve()),
                    "captions": class_name.replace("_"," "),
                })

    output_path = Path(output_path).resolve()
    output_path.parent.mkdir(parents=True, exist_ok=True)
    with open(output_path, "w") as f:
        for entry in entries:
            f.write(json.dumps(entry) + "\n")

    print(f"Wrote {len(entries)} entries to {output_path}")
    return len(entries)

# Run
create_train_json(DATASET_DIR, OUTPUT_FILE, DATASET_NAME)
```
